** # Efficient way to combining multiple csv files 


```python
import pandas as pd 
import glob
file_path = './Resources_CITY_BIKE_ASSIGNMENT/*.csv'
csv_files = glob.glob(file_path)
```

```python
df_list = [pd.read_csv(file) for file in csv_files] 
combined_df = pd.concat(df_list, ignore_index=True)

```

```python
# Convert the columns to datetime 
combined_df['started_at'] = pd.to_datetime(combined_df['started_at']) 
combined_df['ended_at'] = pd.to_datetime(combined_df['ended_at'])

# Extract year and time 
combined_df['year'] = combined_df['started_at'].dt.year 
combined_df['start_time'] = combined_df['started_at'].dt.time 
combined_df['end_time'] = combined_df['ended_at'].dt.time

#Format the time to show two decimal places for seconds 
combined_df['start_time'] = combined_df['started_at'].dt.strftime('%H:%M:%S.%f').str[:-4] 
combined_df['end_time'] = combined_df['ended_at'].dt.strftime('%H:%M:%S.%f').str[:-4]


combined_df
```

```python
#dropping the unwanted columns 
combined_df = combined_df.drop(columns=['started_at','ended_at'])

combined_df
```

```python
combined_df.to_csv('city_bike_data_2022_to_2024.csv', index=False)
```
